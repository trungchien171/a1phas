```python
from ib_insync import *
from IPython.display import display, clear_output
import pandas as pd
import numpy as np
import datetime as dt
util.startLoop()

ib = IB()
ib.connect()   

contract = Forex('EURUSD')

def onBarUpdate(bars, hasnewBar):
    global df 
    df = pd.DataFrame(bars)[['date', 'open', 'high', 'low', 'close', 'volume']].iloc[:-1] # only take the date and closing price, execute the last row
    df.set_index('date', inplace = True) 
    clear_output(wait=True)
    display(df)

bars = ib.reqHistoricalData(
    contract,
    endDateTime='', # current data
    durationStr='10 Y',
    barSizeSetting='1 day',
    whatToShow='MIDPOINT',
    useRTH=True,
    formatDate=2,
    keepUpToDate=True,
    timeout = 0
)
```

```python
bars.updateEvent += onBarUpdate
```

```python
ib.cancelHistoricalData(bars)
```

```python
df.to_csv("daily_eurusd.csv")
```
