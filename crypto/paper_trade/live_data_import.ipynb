```python
import json
import websocket
import pandas as pd
```

```python
endpoint = "wss://stream.binance.com:9443/ws"
our_msg = json.dumps({'method': 'SUBSCRIBE', 'params': ['btcusdt@ticker'], 'id': 1})
```

```python
df = pd.DataFrame()

def on_open(ws):
    ws.send(our_msg)
def on_message(ws, message):
    global df
    out = json.loads(message)
    out = pd.DataFrame({'price': float(out['c'])}, index = [pd.to_datetime(out['E'], unit='ms') + pd.Timedelta(hours=7)])
    df = pd.concat([df, out], axis = 0)
    print(df)
```

```python
ws = websocket.WebSocketApp(endpoint, on_open=on_open, on_message=on_message)
ws.run_forever()
```
