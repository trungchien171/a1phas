## Get all series

```python
# import all forex.csv files from my datasets
def import_all_csv(path):
    files = os.listdir(path)
    csv_files = [f for f in files if f.endswith('.csv')]
    dataframes = []
    for csv_file in csv_files:
        file_path = os.path.join(path, csv_file)
        df = pd.read_csv(file_path)
        df['Date'] = pd.to_datetime(df['Date'])
        dataframes.append(df)

    if dataframes:
        merged_df = pd.merge(dataframes[0], dataframes[1], on='Date', how='outer')
        for df in dataframes[2:]:
            merged_df = pd.merge(merged_df, df, on='Date', how='outer')
    else:
        merged_df = pd.DataFrame()
    return merged_df
```

## Select 3 random series

```python
# randomly select 3 series
def random_subset(df, num_of_cols = 3):
    if len(df.columns) < num_of_cols:
        raise ValueError(f"Dataframe has fewer than {num_of_cols} columns")
    sample_cols = random.sample(list(df.columns), num_of_cols)
    subset = df[sample_cols]
    return subset
```

## Brownian motion test

- In order to test Brownian motion for one series, we must test for its stationary, normality increments, and independence increments.
- I use adfuller test for stationary testing, shapiro for normality test, and acorr_ljungbox for independence test.

```python
from statsmodels.tsa.stattools import adfuller
from scipy.stats import shapiro
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
def test_stationary(series):
    result = adfuller(series)
    print(f'ADF stat: {result[0]}')
    print(f'p_value: {result[1]}')
    for key, value in result[4].items():
        print(f'Critical Value {key}: {value}')

def test_normality(series):
    stat, p_value = shapiro(series)
    print(f'Stat: {stat}')
    print(f'p_value: {p_value}')

def test_independence(series):
    result = acorr_ljungbox(series, lags=[10], return_df=True)
    print(result)

def test_in_chunks(series, chunk_size=1000): # due to memory error, i seperate the data into smaller chunks then test each chunk
    for start in range(0, len(series), chunk_size):
        end = min(start + chunk_size, len(series))
        chunk = series[start:end]
        print(f'\nTesting chunk from {start} to {end}')
        test_stationary(chunk)
        test_normality(chunk)
        test_independence(chunk)
```

```python
testing_cols = ['ret_NZDCHF', 'ret_USDHKD', 'ret_SGDJPY']

for col in testing_cols:
    print(f'\nTesting for {col} in chunks')
    test_in_chunks(subset[col])
```

From the results, the 3 selected series have stationary, non-normality distributed increments, and indepedent increments

## Pair trading

```python
from statsmodels.tsa.stattools import coint

# johansen cointegration test
def find_cointegrated_pairs(data):
    n = data.shape[1] # number of series in the data
    score_matrix = np.zeros((n,n)) # matrix to store cointegration scores
    p_value_matrix = np.ones((n,n)) # matrix to store p values
    keys = data.keys()
    pairs = []

    # loop through each pair
    for i in range(n):
        for j in range(i+1, n):
            S1 = data[keys[i]]
            S2 = data[keys[j]]

            # perform the cointegration test
            result = coint(S1, S2)
            score = result[0]
            pvalue = result[1]

            # store the results
            score_matrix[i, j] = score
            p_value_matrix[i,j] = pvalue

            # if p-value less than 5%, consider them cointegrated and add to pairs list
            if pvalue < 0.05:
                pairs.append((keys[i],keys[j]))
    return score_matrix, p_value_matrix, pairs
```

```python
frac = 0.01
n = len(df)
num_of_rows = int(n*0.01)
pair_trade_df = df.iloc[:num_of_rows]
pair_trade_df = pair_trade_df.dropna()

np.random.seed(42)
_, p_value_matrix, _ = find_cointegrated_pairs(pair_trade_df)

min_pvalue = float('inf')
best_pair = None

for i in range(len(p_value_matrix)):
    for j in range(i+1, len(p_value_matrix)):
        if p_value_matrix[i,j] < min_pvalue:
            min_pvalue = p_value_matrix[i,j]
            best_pair = (pair_trade_df.columns[i], pair_trade_df.columns[j])

print(f'The best pair is: {best_pair} with a p_value of {min_pvalue}')
```

```python
# backtest/simulate the trade
def trade(S1, S2, window1, window2):
    if (window1 == 0) or (window2 == 0):
        return 0
    
    ratios = S1/S2
    ma1 = ratios.rolling(window=window1,
                               center=False).mean()
    ma2 = ratios.rolling(window=window2,
                               center=False).mean()
    std = ratios.rolling(window=window2,
                        center=False).std()
    zscore = (ma1 - ma2)/std
    
    # start with no money and no positions
    money = 0
    countS1 = 0
    countS2 = 0
    for i in range(len(ratios)):
        # short sell if the z-score < -1
        if zscore[i] < -1:
            money += S1[i] - S2[i] * ratios[i]
            countS1 -= 1
            countS2 += ratios[i]
        # long if the z-score > 1
        elif zscore[i] > 1:
            money -= S1[i] - S2[i] * ratios[i]
            countS1 += 1
            countS2 -= ratios[i]
        # neutral positions if the z-score between -0.75 and 0.75
        elif abs(zscore[i]) < 0.75:
            money += S1[i] * countS1 + S2[i] * countS2
            countS1 = 0
            countS2 = 0
    return money
```

```python
trade(df['CADJPY'].iloc[1000:], df['NZDJPY'].iloc[1000:], 60, 5)
```
